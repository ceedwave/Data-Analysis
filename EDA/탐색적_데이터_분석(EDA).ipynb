```python
# 필요한 패키지 설치
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
sns.set(color_codes=True)
%matplotlib inline
```

```python
# 데이터 불러오기
# https://www.kaggle.com/datasets/jessemostipak/hotel-booking-demand
df = pd.read_csv("./hotel_bookings.csv")

# 데이터 샘플 확인
df.head()
```

```python
# 호텔, 리드타임 불러오기
df2 = df[['hotel', 'lead_time']]
df2.head()
```

```python
# 각 컬럼의 속성 및 결측치 확인
df.info()
```

```python
# 각 컬럼의 통계치 확인
# 갯수, 평균, 표준편차, 최빈값, 사분위수, 최대값 수치
df.describe()
```

```python
# 각 컬럼의 왜도 확인 
# 파라미터 미설정 시 string 타입까지 집계하기 때문에 오류가 발생
df.skew(numeric_only=True)
```

```python
# 각 컬럼의 첨도 확인
# 파라미터 미설정 시 string 타입까지 집계하기 때문에 오류가 발생
df.kurtosis(numeric_only=True)
```

```python
# 특정 변수 분포 시각화
# dpi 값을 통해서 그래프 사이즈 조절 
plt.rcParams['figure.dpi'] = 300
sns.distplot(df['lead_time'])
```

```python
# 호텔 구분에 따른 lead_time 분포 차이 시각화
# 두 가지 그래프를 겹쳐 그렸을 때 데이터 분포를 비교하고 상관관계를 확인할 수 있다.

# 데이터의 분포를 바이올린 형태의 그래프로 표현
sns.violinplot(x="hotel", y="lead_time", data=df, inner=None, color=".8")
# 데이터 포인트를 점으로 표시하여 분포를 시각화 하여 표현
sns.stripplot(x="hotel", y="lead_time", data=df, size=1)
```
